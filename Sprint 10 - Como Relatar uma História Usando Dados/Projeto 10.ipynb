O código acima necessita de ajustes pois a coluna 'street_name' está vazia.


O problema com a expressão regular `r'^[^0-9]*'` é que ela não funciona corretamente para extrair nomes de ruas quando o endereço começa com um número.

Por que essa regex está errada?
Vamos analisar cada parte dela:

    - ^ → Indica que a correspondência começa do início da string.
    - [^0-9]* → Captura qualquer sequência de caracteres que NÃO sejam números.
    - [^0-9] significa "qualquer caractere que não seja um número".
    - * significa "zero ou mais repetições" desse padrão.
    
Isso significa que a regex vai pegar tudo até encontrar um número e então parar. Sendo assim, se o endereço começa com um número, a regex não consegue capturar nada.

Exemplo:
Vamos ver como a regex age em diferentes endereços:

```markdown    
    
| Endereço                | O que a regex captura (`r'^[^0-9]*'`) | Problema |
|-------------------------|--------------------------------|-------------------------------------------------------|
| `"123 Rua das Flores"`  | `""` (string vazia)           | ❌ A regex parou antes do número, então não capturou nada. |
| `"456 Avenida Paulista"`| `""` (string vazia)           | ❌ Mesmo problema. |
| `"Rua das Rosas"`       | `"Rua das Rosas"`             | ✅ Funciona porque não há número no começo. |
    
```   

Além disso, re.match() só analisa o início da string, então ele não olha o resto do endereço para encontrar o nome da rua depois do número.

Para ajustar, é necessário fazer alterações na expressão Regex:

    
```python
    
# Extrair o nome da rua da coluna 'address' e criar uma nova coluna 'street_name'
df['street_name'] = df['address'].str.extract(r'(\d+\s+)?([\w\s]+)')[1]

# Visualizar as primeiras linhas para verificar o resultado
print(df[['address', 'street_name']].head())
    
```

Aqui, a regex \d+\s+(.+) funciona corretamente porque:
    
    - \d+ captura o número no início.
    - \s+ captura o espaço depois do número.
    - (.+) captura todo o resto da string, que é o nome da rua.

Caso o endereço não tenha número, a função apenas retorna o endereço original.
    
Por favor realize o ajuste necessário e rode o código do gráfico de dez ruas com o maior número de restaurantes.

````python
#CORREÇÃO DO ITEM 6:

#6 - Coloque os dados dos nomes das ruas da coluna address em uma coluna separada.

# Extrair o nome da rua da coluna 'address' e criar uma nova coluna 'street_name'
rest_df['street_name'] = rest_df['address'].str.extract(r'(\d+\s+)?([\w\s]+)')[1]

# Visualizar as primeiras linhas para verificar o resultado
print(rest_df[['address', 'street_name']].head())
````

````python
#7- Construa um gráfico de dez ruas com o maior número de restaurantes.

# Contar o número de restaurantes por rua
ruas_com_restaurantes = rest_df.groupby('street_name')['object_type'].apply(lambda x: (x == 'Restaurant').sum())

# Selecionar as 10 ruas com maior número de restaurantes
top_10_ruas = ruas_com_restaurantes.sort_values(ascending=False).head(10)

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
top_10_ruas.plot(kind='bar', color='skyblue')

# Adicionar título e rótulos
plt.title('Top 10 Ruas com o Maior Número de Restaurantes')
plt.xlabel('Ruas')
plt.ylabel('Número de Restaurantes')
plt.xticks(rotation=45, ha='right')

# Exibir o gráfico
plt.tight_layout()
plt.show()
````

````python
#8 - Encontre o número de ruas que têm apenas um restaurante.

# Contar o número de restaurantes por rua
restaurantes_por_rua = rest_df.groupby('street_name')['object_type'].apply(lambda x: (x == 'Restaurant').sum())

# Filtrar as ruas com exatamente um restaurante
ruas_com_um_restaurante = restaurantes_por_rua[restaurantes_por_rua == 1]

# Exibir o número de ruas que têm apenas um restaurante
print(f'Número de ruas com apenas um restaurante: {len(ruas_com_um_restaurante)}')

# Exibir as ruas que têm apenas um restaurante
print('Ruas com apenas um restaurante:')
print(ruas_com_um_restaurante)
````

````python
#9 - Para as ruas com muitos restaurantes, olhe para a distribuição de número de assentos. Quais tendências você consegue notar?

# Definir o critério para muitos restaurantes (ruas com 2 ou mais restaurantes)
ruas_com_muitos_restaurantes = restaurantes_por_rua[restaurantes_por_rua >= 2]

# Filtrar os dados apenas para as ruas com muitos restaurantes
dados_muitas_ruas = rest_df[rest_df['street_name'].isin(ruas_com_muitos_restaurantes.index)]

# Visualizar a distribuição do número de assentos nessas ruas
plt.figure(figsize=(10, 6))
plt.hist(dados_muitas_ruas['number'], bins=30, color='skyblue', edgecolor='black')

# Adicionar título e rótulos
plt.title('Distribuição do Número de Assentos nas Ruas com Muitos Restaurantes')
plt.xlabel('Número de Assentos')
plt.ylabel('Frequência')

# Exibir o gráfico
plt.show()

# Análise de tendência: Média do número de assentos nas ruas com muitos restaurantes
media_assentos_muitas_ruas = dados_muitas_ruas.groupby('street_name')['number'].mean()
print('Média de assentos por rua nas ruas com muitos restaurantes:')
print(media_assentos_muitas_ruas)
````
